```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("StudentsPerformance (1).csv")
data
```

# 1. Find out how many males and females participated in the test

```python
sns.countplot(data["gender"])
count = data.groupby("gender").size()
print(count)
plt.show()
```

# 2. What do you think about the students' parental level of education?

```python
sns.countplot(data["parental level of education"])
plt.xticks(rotation = 60)
plt.show()
```

# 3. Who scores the most onaverage for math,reading and writing based on

# ● Gender

```python
math_gender = data.groupby('gender')['math score',"reading score","writing score"].mean()
type(math_gender)
math_gender.reset_index( inplace=True)
print(math_gender)
math_gender.plot(x = "gender", y = ["math score","reading score", "writing score"], kind = "bar")
plt.xticks(rotation = 0)
plt.show()
```

# ● Test preparation course

```python
math_gender = data.groupby("test preparation course")['math score',"reading score","writing score"].mean()
type(math_gender)
math_gender.reset_index( inplace=True)
print(math_gender)
math_gender.plot(x = "test preparation course", y = ["math score","reading score", "writing score"], kind = "bar")
plt.xticks(rotation = 0)
plt.show()
```
