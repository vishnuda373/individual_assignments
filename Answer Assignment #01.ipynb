# Perform meaningful visualizations using the dataset. Bring at least 3 visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris_data = pd.read_excel("iris_master.xls")
```

### Distribution of SL and SW

```python
colors  = ["red", "blue", "orange"]
classification = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
```

```python
for i in range(3):
    x = iris_data[iris_data["Classification"] == classification[i]]
    plt.scatter(x["SL"],x["SW"], c = colors[i], label = classification[i])
plt.xlabel("SL")
plt.ylabel("SW")
plt.legend()
plt.show
```

### Distribution of PL and PW

```python
for i in range(3):
    x = iris_data[iris_data["Classification"] == classification[i]]
    plt.scatter(x["PL"],x["PW"],c = colors[i], label = classification[i])
plt.xlabel("PL")
plt.ylabel("PW")
plt.legend()
plt.show()
```

### Average of SL, SW, PL, PW of respective Classifications

```python
grouped = iris_data.groupby(["Classification"],as_index = False).mean()
columns = list(grouped.columns)
plt.rcParams["figure.figsize"] = (10,8)
grouped.plot(x = "Classification", y = ["SL", "SW", "PL", "PW"], kind = "bar")
plt.xticks(rotation=45, ha='right')
plt.show()
```
