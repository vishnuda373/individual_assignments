```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("cars_data.csv")
data
```

```python
data.isnull().any()
```

```python
sns.countplot(data["Buyer Gender"])
```

```python
largest = data.nlargest(5, ["Sale Price"])
largest
```

```python
plt.rcParams["figure.figsize"] = (12,12)
largest.plot(x = "Model", y = "Sale Price", kind = "bar")
plt.xticks(rotation = 360)
plt.show()
```

```python
smallest = data.nsmallest(5, ["Sale Price"])
smallest
```

```python
plt.rcParams["figure.figsize"] = (12,12)
smallest.plot(x = "Model", y = "Sale Price", kind = "bar")
plt.xticks(rotation = 360)
plt.show()
```
